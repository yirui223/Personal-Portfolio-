# Pandas操作MySQL数据表

```python
# 前提：你的Anaconda要先装：pymysql 和 sqlalchemy 模块。
# 安装方式：pip install 模块名
# 此外，你的电脑还需要安装：mysql数据库
```

```python
"""
UTF-8：国际通用编码。包含全球所有国家的文字（英文、中文、日文、emoji表情等）。在 Python / Pandas 中，默认就是 UTF-8。
GBK：中国本地专用编码。主要针对简体中文设计，不包含日文、韩文、emoji 等。
index_col=0 让第一列充当索引
"""
# 需求：把这些数据写到MYSQL中
data = pd.read_csv('/Users/yiruihe/Desktop/python/machine_learningProject/csv示例文件.csv',encoding='gbk',index_col=0)
data
```

```python
# 导包
# from sqlalchemy import create_engine # 导入引擎对象 写到第一个单元格了
```

```python
# 创建引擎对象
#                      数据库+模块名：//数据库用户名user：密码@主机名host:端口号/要操作的数据库？编码方式
engine = create_engine('mysql+pymysql://root:password@localhost:3306/Test3?charset=utf8')

# 具体的往数据库写数据的动作
# 参1：数据表名 参2：引擎对象 参3：是否把索引写入数据库， 参4：如果数据表存在，则如何处理
data.to_sql('my_table',engine,index=False,if_exists='replace')
```

```python
# 从SQL表中读取数据
# 参1：数据表名， 参2： 引擎对象
sql_df = pd.read_sql('my_table',engine) #读取全表
sql_df = pd.read_sql('select name, AKA from my_table limit 0,2',engine) # 读取表中指定数据
sql_df
```
