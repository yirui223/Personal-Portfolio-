# Series对象入门

#### 创建Series对象

```python
import pandas as pd
import numpy as np
```

```python
# 创建Series对象，采用：默认自增索引
s1 = pd.Series([1,2,3,4,5])
print(s1)
```

```python
# 创建Series对象，采用：自定义索引
s2 = pd.Series([1,2,3,4,5],index = ['a','b','c','d','e'])
print(s2)
```

```python
# 使用字典，元组创建Series对象
# 元组形式
s3 = pd.Series((11,22,33,44,55))
print(s3)
# 字典形式
s4 = pd.Series({'a':1,'b':2,'c':3,'d':4})
print(s4)
```

```python
# 使用numpy > 创建Series对象
S5 = pd.Series(np.arange(5))
print(S5)
```

# DataFrame对象入门

         - · 每一列就是一个 Series（带有相同索引的一维数组）。
         - · 所有列共享同一个行索引（index），拼在一起就成了 DataFrame。

```python
# 场景1 通过 字典 + 列表 的方式实现
# 准备数据集，每个键值对等于 一列数据
info = {
    'name': ['a','b','c','d'],
    'gender': ['male','female','male','female'],
    'age':[23,56,94,55]
}
# 把上述的数据集， 封装成DataFrame对象
df1 = pd.DataFrame(data = info)
print(df1)
```

```python
# 场景1 通过 列表 + 元组 的方式实现
# 准备数据集，每个元组对等于 一行数据
info = [
    ('a','女',39),
    ('b','女',30),
    ('c','女',32),
]
# 把上述的数据集， 封装成DataFrame对象
df2 = pd.DataFrame(data = info)
print(df2)
```

```python
# 场景1 通过 列表 + 元组 的方式实现
# 准备数据集，每个元组对等于 一行数据
info = [
    ('a','女',39),
    ('b','女',30),
    ('c','女',32),
]
# 把上述的数据集， 封装成DataFrame对象
df2 = pd.DataFrame(data = info,columns=['name','gender','age'])
print(df2)
```

```python
# 场景1 通过 列表 + 列表 的方式实现
# 准备数据集，每个列表对等于 一行数据
info = [
    ['a','女',39],
    ['b','女',30],
    ['c','女',32],
]
# 把上述的数据集， 封装成DataFrame对象
df2 = pd.DataFrame(data = info,columns=['name','gender','age'])
print(df2)
```

```python
# 通过 numpy 的 ndarray 变成 pandas DataFrame 的方式实现
# 创建numpy 的 ndarray 对象
arr1 = np.arange(12).reshape(3,4)
print(arr1)
```

```python
# 把上述的ndarray 对象，封装成DataFrame对象
df3 = pd.DataFrame(data = arr1,columns=['a','b','c','d'])
print(df3)
```

# DataFrame建表需要注意的细节

```python
'''
关键区别：

· 单层 []：列表里的每个元素都变成一行。
· 双层 [[]]：内层列表整体变成一行，内层列表里的每个元素变成一列。

在 pandas 的 DataFrame 里，元组和列表作为数据容器时，建出来的东西没有区别。
'''
import pandas as pd

# 情况1：双层括号 → 里面有3列 → 1行3列
df1 = pd.DataFrame([['A', 'B', 'C']])
print(df1)
#     0  1  2
# 0   A  B  C
print(df1.shape)  # (1, 3) → 1行3列


# 情况2：单层括号 → 只有1列 → 3行1列
df2 = pd.DataFrame(['A', 'B', 'C'])
print(df2)
#     0
# 0   A
# 1   B
# 2   C
print(df2.shape)  # (3, 1) → 3行1列
```

```python
# 元组和列表效果一样
df1 = pd.DataFrame([('a', 'b', 'c')])  # 双层元组 → 1行3列
df2 = pd.DataFrame(['a', 'b', 'c'])    # 单层列表 → 3行1列
df3 = pd.DataFrame(('a', 'b', 'c'))    # 单层元组 → 3行1列
print(df1)
print('-' * 20)
print(df2)
print('-' * 20)
print(df3)
```

```python
# 字典 → 每个键值对是一列
df = pd.DataFrame({
    'name': ['A', 'B', 'C'],   # 这一列叫 name
    'age': [25, 30, 35]        # 这一列叫 age
})
# 结果：3行2列，有列名
print(df)
```

#### apply()函数 -> 它可以执行自定义函数

```python
# 背景：目前我们用的都是Pandas提供好的函数，例如：count(), max(), min() ...
# 如果要传入一些自定义的逻辑，此时就需要用到 apply()函数了
# 格式：df.apply(自定义函数, axis=0)  # 0 → 列, 1 → 行
```

```python
# 需求：同时获取到多列的最大值和最小值的差值。例如：open列，close列
df.open.max() - df.open.min()
df.close.max() - df.close.min()
# 1. 自定义函数 my_func，接收某列的数据，计算该列的最大值和最小值，返回差值。
def my_func(col):
    return col.max() - col.min()

# 2. 获取到 open列 和 close 列
df[['open','close']]
df.loc[:,['open','close']]
df.iloc[:,[0,2]]
df.iloc[:,0:3:2]

# 3. 通过apply()函数，调用上述的自定义函数，作用到指定的列。
df[['open','close']].apply(my_func)    # 把df的各个列分别传入函数，计算结果
```

```python
df[['open','close']].apply(my_func,axis = 1) # axis = 1 表示 对 行 进行操作
```

```python
df[['open','close']].apply(my_func,axis = 0) # axis = 0 表示 对 列 进行操作  默认是0
```

```python
# 思路2：合并版，通过 lambda函数实现
df[['open','close']].apply(lambda col:col.max()-col.min())
```
