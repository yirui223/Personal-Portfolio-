## 创建Numpy的ndarray对象

- array()函数，把Python列表 变成 ndarray对象

```python
# 创建Python列表对象
my_list = [11,22,33,44,55]
print(type(my_list))

#把python列表 变成 ndarray
arr2 = np.array(my_list)
print(arr2)
print(type(arr2))
```

- arange()函数，创建ndarray对象

```python
# arange(起始，结束，步长，dtype=类型)，类似于 Python中的range()函数。
# 1. 创建ndarray对象。
arr3 = np.arange(0, 10, 2, dtype=np.int32)   #相当于可以自定义
# 2. 打印ndarray对象。
print(arr3)
print(type(arr3))
print(f'numpy的元素类型{arr3.dtype}')
```

- 随机数生成ndarray对象

```python
# random.rand() 生成 0.0~1.0 之间， 包左不包右的 随机小数矩阵
arr4 = np.random.rand(3,5)    #3行5列
print(arr4)
print(type(arr4))
```

```python
# random.randint(起始值，结束值，size = (行，列）) 生成 指定范围， 包左不包右的 随机整数矩阵
arr5 = np.random.randint(3,9,size=(2,6))    #2行6列
print(arr5)
print(type(arr5))
```

```python
# random.uniform(起始值，结束值，size = (行，列）) 生成 指定范围， 包左不包右的 随机小数矩阵
arr6 = np.random.uniform(3,9,size=(2,6))
print(arr6)
print(type(arr6))
```

## 创建等比数列

```python
# 1. 创建等比数列。 参1：起始值（充当10的指数），参2：结束值（充当10的指数），参三：元素个数, 包左包右
arr9 = np.logspace(0,3,4)
print(arr9)
print(arr9.dtype)
```

```python
# 1. 创建等比数列。 参1：起始值（充当2的指数），参2：结束值（充当2的指数），参三：元素个数, 包左包右, 参四：修改底数为2,参五：修改数值类型
arr10 = np.logspace(0,3,4,base = 2,dtype=np.int64)
print(arr10)
print(arr10.dtype)
```

## 创建等差数列

```python
# 1. 创建等差数列
arr11 = np.linspace(1,10,4)
print(arr11)
print(arr11.dtype)
```

```python
# 1. 创建等差数列
arr11 = np.linspace(1,10,4,endpoint = False)
print(arr11)
print(arr11.dtype)
```

```python
# 1. 创建等差数列参1：起始值，参2：结束值，参三：元素个数, 包左包右, 参四：是否包含结束值（包左不包右）,参五：修改数值类型
arr12 = np.linspace(1,10,4,endpoint = False,dtype=np.int64)
print(arr12)
print(arr12.dtype)
```

# 排序函数

```python
# 通过 np.sort() 方式排序 > 返回新副本
arr17 = np.array([11,33,22,66,55])
print(arr17)
arr18 = np.sort(arr17)
print(arr18)
```

# Numpy的矩阵运算

- 减法运算

```python
# 定义数组
arr1 = np.array([10,20,30,40])
arr2 = np.array(4)
print(arr1)
print(arr2)
# 减法运算
print(arr1 - arr2)   #对应元素做减法运算
# 加法运算
print(arr1 + arr2)
```

- 矩阵乘法

```python
# 1. 定义数组。
arr1 = np.array([[1, 2, 3], [4, 5, 6]])
arr2 = np.array([[1, 2, 3], [4, 5, 5]])
# 2. 打印数组
print(f'arr1: {arr1}')
print(f'arr2: {arr2}')
```

```python
# 3. 矩阵乘法，行列数一致。
print(arr1 * arr2)  # 对应位置(坐标)的元素做乘法运算。

print(np.multiply(arr1, arr2))   # 效果同上
```

```python
# 4. 矩阵乘法，行列数不一致。
arr3 = np.array([[1, 2, 3], [4, 5, 6]])
arr4 = np.array([[9, 8], [6, 5], [3, 2]])
# 5. 打印数组。
print(f'arr3: {arr3}')
print(f'arr4: {arr4}')
```

```python
# 6. 矩阵乘法 → 行列数不一致。
# print(arr3 * arr4)    # 报错，行列数不一致。

print(arr3 @ arr4)       #语法糖      A：2行3列     B：3行2列      A列 = B行 即可操作    结果是：A行B列
print(np.dot(arr3, arr4))
print(arr3.dot(arr4))
```

```python
# 7. A列 = B行即可操作，结果是：A行B列
arr3 = np.array([[1, 2, 3, 4], [4, 5, 6, 7]])  # 2行4列
arr4 = np.array([[9, 8], [6, 5], [3, 2]])  # 3行2列
print(arr3 @ arr4)    # error
```
